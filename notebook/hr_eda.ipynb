Data Cleaning

```python
# Check Missing Values
df.isnull().sum()
```

```python
# Drop Duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# Convert dtype of Joining Date
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
```

Feature Engineering

```python
# Extract Joining Year
df["Joining_Date"] = df["Joining_Date"].dt.year
```

```python
# Salary in Lakhs
df["Salary_LPA"] = (df["Salary"] * 12) / 100000
df["Salary_LPA"] = df["Salary_LPA"].round(2)
```

Data Visualization

```python
# Employees by Department
df["Department"].value_counts().plot(kind="bar")
plt.title("Employees by Department")
plt.show()
```

```python
# Gender Distribution
df["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
# Salary Distribution
plt.hist(df["Salary"], bins=20)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Employees")
plt.show()
```

```python
# Attrition
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Employee Attrition")
plt.show()
```

```python
# 5. Department-wise Average Salary
df.groupby("Department")["Salary"].mean().plot(kind="bar")
plt.title("Average Salary by Department")
plt.show()
```

```python
# Performance Ratings
df["Performance_Rating"].value_counts().sort_index().plot(kind="bar")
plt.title("Performance Ratings")
plt.show()
```
